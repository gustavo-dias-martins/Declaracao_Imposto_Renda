```python
#criamos as funções que serão usadas ao longo do projeto
def preco_total(lista_preco_unitario,lista_quantidade):
    
    ''''
    Realizar o preço total de um DataFrame de uma ação 
    
    Parâmetros:
    lista_preco_unitario : a lista de preços de compra ou venda daquela ação
    lista_quantidade : a lista de quantidade da operação daquela ação
    '''
    
    preco_total = sum(lista_preco_unitario * lista_quantidade )
    return preco_total

def diferenca_quantidades(lista_compra_quantidade,lista_venda_quantidade):
    
    ''''
    Realizar a diferença do total de quantidades de um DataFrame de uma ação 
    
    Parâmetros:
    lista_compra_quantidade : a lista de quantidades de compra daquela ação
    lista_venda_quantidade : a lista de quantidades de venda daquela ação
    '''
    
    diferenca = sum(lista_compra_quantidade) - sum(lista_venda_quantidade)
    return diferenca

def conversao_numero_brasileiro(num):
        
    ''''
    Realizar a conversão de ponto em vírgula e de vírgula em ponto
    
    Parâmetros:
    num : o número que será convertido
    '''
    return num.replace(',','_').replace('.',',').replace('_','.')
```

```python
bases_importadas = bases_importadas.dropna(axis =1)

#fazer uma lista única com todos os tickers das AÇÕES e FIIS
lista_tickers =bases_importadas['Código de Negociação'].unique()
```

```python
#criar um DataFrame final inicialmente vazio para incluir as informações finais dos tickers
base_final =pd.DataFrame(columns=['Código de Negociação','Quantidade','Preço Ponderado'])

#Passar por cada ticker, separar em compra e venda e realizar as operações da media_ponderada e qtd total final
for ticker in lista_tickers:
    base_compra = bases_importadas.loc[(bases_importadas['Código de Negociação'] == f'{ticker}') & (bases_importadas['Tipo de Movimentação'] == 'Compra')]
    base_venda = bases_importadas.loc[(bases_importadas['Código de Negociação'] == f'{ticker}') & (bases_importadas['Tipo de Movimentação'] == 'Venda')]
    preco_total_compra = preco_total(base_compra['Preço'], base_compra['Quantidade'])
    preco_total_venda = preco_total(base_venda['Preço'], base_venda['Quantidade'])
    qtd = diferenca_quantidades(base_compra['Quantidade'],base_venda['Quantidade'])
    
#se desfizemos daquele ticker, a qtd é ZERO e devemos pular 
    if qtd == 0:
        continue
    else:
        media_ponderada =(preco_total_compra - preco_total_venda) / qtd
        
#criar um DataFrame intermediario para armazenar o ticker e depois adicionar a linha dele no DataFrame base_final
        base_exportar= pd.DataFrame([[f'{ticker}',qtd,round(media_ponderada,3)]],columns=['Código de Negociação','Quantidade','Preço Ponderado'])
        base_final = base_final.append(base_exportar)
base_final['Quantidade'] = pd.to_numeric(base_final['Quantidade'], errors= 'coerce')
```

```python
#valores declarados no ano anterior
declaracao_ano_anterior = declaracao_ano_anterior[['Ticker','Valor Total']]
declaracao_ano_anterior = declaracao_ano_anterior.rename(columns={'Valor Total':'Valor Ano Anterior'})
display(declaracao_ano_anterior)
```

```python
#Juntar a base final com as informações dos fundos
base_final['Valor Total'] = base_final['Quantidade'] * base_final['Preço Ponderado']
base_final = base_final.rename(columns={'Código de Negociação':'Ticker'})
base_final = base_final.merge(declaracao_ano_anterior, how = 'right')
base_final = base_final.merge(info_tickers_fundo, how='outer')
base_final = base_final.drop(['CNPJ Desarrumado'], axis = 1)
base_final = base_final.sort_values('Ticker')
base_final = base_final.reset_index(drop=True)

display(base_final)
```
